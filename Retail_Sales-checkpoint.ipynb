```python
df['Month_Year'] = df['Order_Date'].dt.to_period('M')
```

```python
category_sales = df.groupby('Product_Category')['Total_Sales'].sum().sort_values(ascending=False)
print(category_sales)
```

```python
plt.figure(figsize=(8,5))
category_sales.plot(kind='bar', color='skyblue')
plt.title('Sales by Product Category')
plt.ylabel('Total Sales')
plt.show()
```

```python
top_products = df.groupby('Product_Name')['Total_Sales'].sum().sort_values(ascending=False).head(10)
print(top_products)
```

```python
plt.figure(figsize=(8,5))
top_products.plot(kind='barh', color='orange')
plt.title('Top 10 Products by Sales')
plt.xlabel('Total Sales')
plt.gca().invert_yaxis()  # Highest on top
plt.show()
```

```python
top_customers = df.groupby('Customer_Name')['Total_Sales'].sum().sort_values(ascending=False).head(10)
print(top_customers)
```

```python
plt.figure(figsize=(8,5))
top_customers.plot(kind='barh', color='green')
plt.title('Top 10 Customers by Revenue')
plt.xlabel('Total Sales')
plt.gca().invert_yaxis()
plt.show()
```

```python
payment_sales = df.groupby('Payment_Method')['Total_Sales'].sum().sort_values(ascending=False)
print(payment_sales)
```

```python
plt.figure(figsize=(6,4))
payment_sales.plot(kind='bar', color=['orange','green','blue'])
plt.title('Total Sales by Payment Method')
plt.ylabel('Total Sales')
plt.xlabel('Payment Method')
plt.show()
```

```python
payment_counts = df['Payment_Method'].value_counts()
print(payment_counts)
```

```python
plt.figure(figsize=(6,6))
payment_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['orange','green','blue'])
plt.title('Payment Method Usage (Orders)')
plt.ylabel('')
plt.show()
```
